```python
company_data["Sales"] = pd.cut(company_data["Sales"], bins =[0,8,17],labels = ["Risky","Good"])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
company_data['Sales']= label_encoder.fit_transform(company_data['Sales']) 
company_data['US']= label_encoder.fit_transform(company_data['US'])
company_data['ShelveLoc']= label_encoder.fit_transform(company_data['ShelveLoc']) 
company_data['Urban']= label_encoder.fit_transform(company_data['Urban']) 
```

```python
company_data
```

```python
company_data['Sales'].nunique()
```

```python
company_data.dtypes
```

```python
x = company_data.drop(labels='Sales',axis=1)
y = company_data[['Sales']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(16,8))
plot_tree(model,filled=True,fontsize=10,rounded=True)
plt.show()
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('accuracy score: ', accuracy_score(y_test,y_pred))
print('confusion matrix: \n',confusion_matrix(y_test,y_pred))
print('classification report: \n',classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(estimator=model,param_grid={'criterion': ['gini','entropy'],
                                                'max_depth':[4,5,6,7,8,9]},
                                                cv=5)
grid_search.fit(x,y)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
from sklearn.model_selection import cross_val_score 
cvs= cross_val_score(estimator=model,X=x,y=y,cv=5)
cvs
```

```python
model_2 = DecisionTreeClassifier(criterion = 'entropy',max_depth=9)
model_2.fit(x_train,y_train)
y_pred2 = model_2.predict(x_test)
```

```python
print('accuracy score: ', accuracy_score(y_test,y_pred2))
print('confusion matrix: \n',confusion_matrix(y_test,y_pred2))
print('classification report: \n',classification_report(y_test,y_pred2))
```
