```python
from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
```

```python
Fraud_data['Undergrad']=Le.fit_transform(Fraud_data['Undergrad'])
Fraud_data['Marital.Status']=Le.fit_transform(Fraud_data['Marital.Status'])
Fraud_data['Urban']=Le.fit_transform(Fraud_data['Urban'])
```

```python
Fraud_data
```

```python
Fraud_data.dtypes
```

```python
Fraud_data['Marital.Status'].nunique()
```

```python
Fraud_data['Risky']=''
Fraud_data.loc[Fraud_data['Taxable.Income']<=30000,'Risky']='1'
Fraud_data.loc[Fraud_data['Taxable.Income']>30000,'Risky']='0'
```

```python
Fraud_data
```

```python
del Fraud_data['Taxable.Income']
```

```python
Fraud_data
```

```python
Fraud_data['Risky'].unique()
```

```python
x = Fraud_data.drop(labels='Risky',axis=1)
y = Fraud_data['Risky']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2)
```

```python
x_train.shape, y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,8))
plot_tree(model,filled=True,fontsize=10,rounded=True)
plt.show()
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print('accuracy score: ', accuracy_score(y_test,y_pred))
print('confusion matrix: \n',confusion_matrix(y_test,y_pred))
print('classification report: \n',classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(estimator=model,param_grid={'criterion': ['gini','entropy',''],
                                                'max_depth':[4,5,6,7,8,9],
                                                'splitter' : ['best', 'random']},
                                                cv=5)
grid_search.fit(x,y)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
from sklearn.model_selection import cross_val_score 
cvs= cross_val_score(estimator=model,X=x,y=y,cv=3)
cvs
```

```python
model_2 = DecisionTreeClassifier(criterion = 'entropy',max_depth=3,)
model_2.fit(x_train,y_train)
y_pred2 = model_2.predict(x_test)
```

```python
print('accuracy score: ', accuracy_score(y_test,y_pred2))
print('confusion matrix: \n',confusion_matrix(y_test,y_pred2))
print('classification report: \n',classification_report(y_test,y_pred2))
```
